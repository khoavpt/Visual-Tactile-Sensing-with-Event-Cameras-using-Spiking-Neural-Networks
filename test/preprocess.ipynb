```python
import dv_processing as dv
import pandas as pd

# Open a file
reader = dv.io.MonoCameraRecording("../data/raw_data/raw6.aedat4")

# Get and print the camera name
print(f"Opened an AEDAT4 file which contains data from [{reader.getCameraName()}] camera")

# Initialize an empty list to store event data
events_list = []

# Run the loop while camera is still connected
while reader.isRunning():
    # Read batch of events
    events = reader.getNextEventBatch()
    
    if events is not None:
        events_list.append({
                "start": events.getLowestTime(),
                "end": events.getHighestTime(),
                "events": len(events)
            })
df = pd.DataFrame(events_list)
df['start'] = df['start'] - df['start'].min()
df['end'] = df['end'] - df['end'].min()
save_path = "events.csv"
df.to_csv(save_path, index=False)
```
